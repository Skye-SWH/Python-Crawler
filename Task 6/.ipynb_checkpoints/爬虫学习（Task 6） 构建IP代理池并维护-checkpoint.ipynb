```python
###爬取10页IP代码，存入代理池###

data = collections.defaultdict(list)
url = 'https://www.xicidaili.com/nn'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'}
time_count = 5
count = 0
pages = 10

for page in range(1,pages+1):
    url_new = 'https://www.xicidaili.com/nn/{}'.format(page)
    r = requests.get(url, headers = headers, proxies = proxies)
    r.encoding = 'utf-8'
    soup = BeautifulSoup(r.text,'html.parser')
    trs = soup.find_all('tr')
    for i in range(1,len(trs)):
        tr = trs[i]
        tds = tr.find_all('td')
        IP2 = tds[1].text
        IP3 = tds[2].text
        IP1 = tds[5].text.lower()
        IP = IP1 + '://' + IP2 + ':' + IP3
        print(IP)
        data['IP'].append(IP)
        count += 1
    time.sleep(time_count)
    
    if count%5 == 0:
        print ('已经爬取%d条IP'%count)

###保存数据
df = pd.DataFrame(data)
df 
# df.to_excel('URL数据爬取.xlsx', index=False)
```

```python
###测试代理池的Ip，通过的留下，未通过的剔除###

url_test = 'https://www.xicidaili.com'
count_fail = 0
for IP in data['IP']:
#     print(IP.split('://')[0])
    IP_header = IP.split('://')[0]
    proxies = {IP_header:IP}
#     print(proxies)
    
    try:
        response = requests.get(url_test,headers=headers,proxies=proxies,timeout=10)
        if response.status_code == 200:
#             ip = {'ip':ip_for_test}
            print('{}测试通过'.format(IP))
        else:
            data['IP'].remove(IP)
            count_fail += 1 
            print('{}测试不通过，剔除'.format(IP))
    except Exception as e:
        continue

size = len(data['IP'])
print('{}个IP测试不同过，IP代理池{}个有效'.format(count_fail,size))
```
