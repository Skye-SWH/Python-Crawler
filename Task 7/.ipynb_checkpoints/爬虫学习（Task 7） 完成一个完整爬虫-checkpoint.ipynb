#### 爬取东方股吧内容+评论

```python
########获取所有分页的文章内容#######
import requests
from bs4 import BeautifulSoup
import collections
import pandas as pd

data = collections.defaultdict(list)

pages = 10
for page in range(1,pages+1):
    pages_url = 'http://guba.eastmoney.com/list,000673_{}.html'.format(str(page))
#    print (pages_url)
    res = requests.get(pages_url)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text,'html.parser')
    for news in soup.select('.articleh '):
        news_url = news.select('.l3')[0].select('a')[0]['href']
        if len(news_url) == 27 or len(news_url) == 25:
            news_url_full =  'http://guba.eastmoney.com'+news_url
    #        print(news_url_full)
            res = requests.get(news_url_full)
            res.encoding = 'utf-8'
            soup = BeautifulSoup(res.text,'html.parser')
            content = soup.select('.stockcodec')[0].text.strip()
            if content != '':
                name = soup.select('#zwconttb')[0].select('a')[0].text
                time = soup.select('.zwfbtime')[0].text.split(' ')[1]
                title = soup.select('#zwconttbt')[0].text.strip()
        #        print(content)
                data['time'].append(time)
                data['URL'].append(news_url_full)
                data['name'].append(name)
                data['title'].append(title)
                data['content'].append(content)
                
###保存数据
new_df = pd.DataFrame(data)
new_df.sort_values(by=['time'], ascending=[False], inplace=True)
new_df.to_excel('东方股吧.xlsx', index=False)
```

```python
######获取东方股吧用户评论#########
import requests
from bs4 import BeautifulSoup
import collections
import pandas as pd


data = collections.defaultdict(list)

pages = 2
for page in range(1,pages+1):
    pages_url = 'http://guba.eastmoney.com/list,000673_{}.html'.format(str(page))
#    print (pages_url)
    res = requests.get(pages_url)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text,'html.parser')
    for news in soup.select('.articleh '):
        news_url = news.select('.l3')[0].select('a')[0]['href']
        if len(news_url) == 27 or len(news_url) == 25:
            news_url_full =  'http://guba.eastmoney.com'+news_url
            #for news_url_full_get in news_url_full:
            #news_url_full='http://guba.eastmoney.com/news,000673,795308012.html'
            res = requests.get(news_url_full)
            res.encoding = 'utf-8'
            soup = BeautifulSoup(res.text,'html.parser')
            for p in soup.select('.zwlitxt'):
                user_name = p.select('.zwnick')[0].text 
                user_time = p.select('.zwlitime')[0].text.split()[1]
                user_comment = p.select('.short_text')[0].text.strip()
               # print (user_comment)
                data['URL'].append(news_url_full)
                data['user_name'].append(user_name)
                data['user_time'].append(user_time)
                data['user_comment'].append(user_comment)

new_df = pd.DataFrame(data)
new_df.sort_values(by=['user_time'], ascending=[False], inplace=True)
new_df.to_excel('东方股吧用户评论.xlsx', index=False)
######获取东方股吧用户评论#########
```
