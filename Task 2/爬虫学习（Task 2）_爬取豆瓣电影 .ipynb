```python
# 导入需要的库

import requests
import re
import collections
import pandas as pd 
```

#### 获得每页电影的名次、影片名称、国家、导演等内容

```python
# 获取10个页面URL

for i in range(10):
    url = 'https://movie.douban.com/top250?start={}&filter='.format(i*25)
    print(url)
```

```python
data = collections.defaultdict(list)

###内容爬取
for i in range(10):
    headers = { 'User-Agent':"Mozilla/5.0 (Macintosh; U; Intel Mac OS X 10_6_8; en-us) AppleWebKit/534.50 (KHTML, like Gecko) Version/5.1 Safari/534.50"}
    url = 'https://movie.douban.com/top250?start={}&filter='.format(i*25)
    res = requests.get(url, headers=headers)
    ranks = re.findall(' <em class="">(.*?)</em>',res.text, re.S) #排名
    names = re.findall('<span class="title">([\u4e00-\u9fa5]+)</span>',res.text, re.S) #电影名
    countries = re.findall('&nbsp;/&nbsp;([\u4e00-\u9fa5]+)&nbsp;/&nbsp;', res.text, re.S) #国家
    d = re.sub('导演: ',"",res.text)  
    directors = re.findall('<p class="">(.*?)&nbsp;&nbsp;', d, re.S) #导演
    scores = re.findall('<span class="rating_num" property="v:average">(.*?)</span>',res.text,re.S) #评分

    for rank,name,country,director,score in zip(ranks,names,countries,directors,scores):
        data['Rank'].append(rank)
        data['Moviename'].append(name)
        data['Country'].append(country)
        data['Director'].append(director.strip()) 
        data['Score'].append(score)

data
```

```python
###数据保存到excel

new_df = pd.DataFrame(data)
new_df.to_excel('豆瓣电影Top250信息爬取.xlsx', index=False)
```
